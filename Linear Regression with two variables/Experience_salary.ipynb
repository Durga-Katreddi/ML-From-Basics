```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('hiring.csv')
data
```

```python
#Data Cleaning
data.experience = data.experience.fillna('zero')
median = data['test_score(out of 10)'].median()
data['test_score(out of 10)'] = data['test_score(out of 10)'].fillna(median)
data
```

```python
import word2number.w2n as w2n
data['experience'] = data['experience'].apply(w2n.word_to_num)
data
```

```python
%matplotlib inline
plt.xlabel('experience')
plt.ylabel('salary($)')
plt.scatter(data.experience,data['salary($)'],color='red',marker='+')
```

```python

print(data.columns)
reg = linear_model.LinearRegression()
reg.fit(data[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], data['salary($)'])
```
